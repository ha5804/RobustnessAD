```python
from pathlib import Path
from sklearn.metrics import roc_auc_score
import pandas as pd

Root = Path("/Users/taehayeong/Desktop/PILAB/RobustnessAD/results/winclip")

classes = [
    "bottle", "cable", "capsule", "hazelnut", "metal_nut",
    "pill", "screw", "toothbrush", "transistor", "zipper",
    "wood", "tile", "grid", "leather", "carpet"
]

rows = []

for cls in classes:
    csv_path = Root / f"{cls}.csv"
    data = pd.read_csv(csv_path)

    score = roc_auc_score(data["label"], data["image_score"])
    rows.append({
        "class": cls,
        "I-AUROC": score
    })

df = pd.DataFrame(rows)
df.to_csv(Root / "auroc_mvtec.csv", index=False)
df = df.sort_values(by="I-AUROC", ascending=True)
df
```
